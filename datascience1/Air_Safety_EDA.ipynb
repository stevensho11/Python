```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/stevensho11/BCS109/master/datascience1/airline-safety.csv")
df.head()
```

```
#check for nulls
df.isnull().sum()
```

```
#list number of unique values
df.nunique()
```

```
#stats on my data
df.describe()
```

```
#check scale
sns.catplot(data=df, kind='box', height=3, aspect=4.1)
```

```
#barplot - airlines and the number of incidents had during 2000-2014
bp = sns.catplot(x = 'incidents_00_14', y = 'airline', data =df, kind='bar', height = 7.8, aspect=2.1)
```

```
#plots
sns.catplot(x='incidents_00_14', data=df, kind='count', height=3, aspect=2.1)
sns.catplot(x='fatalities_00_14', data=df, kind='count', height=3, aspect=2.5)
```

```
#histograms
df.hist(figsize=(20, 20), bins=10, xlabelsize=10, ylabelsize=10)
```

```
#swarmplot - avail seats per week vs # incidents
swarmdf = sns.swarmplot(x="incidents_00_14", y="avail_seat_km_per_week", data=df)
```
